```python
import os
root=os.getcwd()
filename=os.path.join(root,'company_sales_data.csv')
filename
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv(filename)
data.head()
```

```python
month=data['month_number'].tolist()
facecream=data['facecream'].tolist()
facewash=data['facewash'].tolist()
toothpaste=data['toothpaste'].tolist()
soap=data['bathingsoap'].tolist()
shampoo=data['shampoo'].tolist()
moisturizer=data['moisturizer'].tolist()
units=data['total_units'].tolist()
profit=data['total_profit'].tolist()
y_label=[i*1000 for i in range(1,16)]
```

```python
plt.figure(figsize=(7,5))
plt.title("Количество проданных средств")
plt.plot(month,facecream,label ='Крем для лица',marker='o')
plt.plot(month,facewash,label = 'Средство для умывания',marker='o')
plt.plot(month,toothpaste,label = 'Зубная паста',marker='o')
plt.plot(month,soap,label = 'Мыло',marker='o')
plt.plot(month,shampoo,label = 'Шампунь',marker='o')
plt.plot(month,moisturizer,label = 'Увлажняющий крем',marker='o')
plt.legend(loc = 'upper left')
plt.xlabel('Месяцы')
plt.ylabel('Количество штук')
plt.xticks(month)
plt.yticks(y_label)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(7,6))
soap_bar=plt.bar(month,soap,zorder=2, color='c')
plt.grid('on', zorder=1)
plt.title("Количество проданного мыла")
plt.xlabel('Месяцы')
plt.ylabel('Количество штук')
plt.xticks(month,rotation='45')
plt.yticks(y_label) 
plt.bar_label(soap_bar, label_type='edge')
plt.show()         
```

```python

facecream_sales=sum(facecream)
facewash_sales=sum(facewash)
toothpaste_sales=sum(toothpaste)
soap_sales=sum(soap)
shampoo_sales=sum(soap)
moisturizer_sales=sum(moisturizer)
product_sales=[facecream_sales,facewash_sales,toothpaste_sales,soap_sales,shampoo_sales,moisturizer_sales]
names=['Крем для лица','Средство для умывания','Зубная паста','Мыло','Шампунь','Увлажняющий крем']
```

```python
plt.pie(product_sales,labels=names,wedgeprops=dict(width=0.8),shadow=True,autopct='%0.1f%%')
plt.title("Структура продаж за год")
plt.show()
```
